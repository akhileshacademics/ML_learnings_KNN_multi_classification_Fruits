```python
import pandas as pd

data = {
    "fruit_label": [1, 1, 1, 2, 2],
    "fruit_name": ["apple", "apple", "apple", "mandarin", "mandarin"],
    "mass": [192, 180, 176, 86, 84],
    "width": [8.4, 8.0, 7.4, 6.2, 6.0],
    "height": [7.3, 6.8, 7.2, 4.7, 4.6],
    "color_score": [0.55, 0.59, 0.60, 0.80, 0.79]
}

df = pd.read_csv("fruit.csv")
df = df.drop(columns=["fruit_subtype"])
df = df.drop(columns=["color_score"])
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

X=df[["mass","height","width"]]
Y=df["fruit_label"]
X
```

```python
Y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=.20,random_state=0)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski')\

knn.fit(X_train,y_train)
```

```python
import numpy as np

y_pred=knn.predict(X_test)

y_pred
```
